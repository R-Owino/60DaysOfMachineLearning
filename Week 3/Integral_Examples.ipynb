### Integral Rules

There a several integral rules but we'll focus on the following, with worked examples:

#### 1. Power rule

It states that:
$$
\int f(x)^n \, dx = \frac{x^{n+1}}{n+1} + C
$$

#### 2. Constant Rule

The integral of 1, a constant is $x$
$$
\int 1 \, dx = x + C
$$

#### 3. Integral of $e^x$
The integral of $e^x$ is $e^x$ itself, because the derivative of $e^x$ is still $e^x$
$$
\int e^x \, dx = e^x + C
$$

#### 4. Integral of $a^x$
As long as $a \neq 1$, the integral of $a^x$ involves dividing by the natural logarithm of the base $a$. This is because the derivative of $\frac{a^z}{ln \; a}$ with respect to $x$ is $a^x$
$$
\int a^x dx = \frac{a^z}{ln \; a} + C
$$

#### 2. Definite Integration problems

##### Example 1: Power Rule
1. Evaluate $\int_{1}^{3} x^2 \, dx$
$$
\int_{1}^{3} x^2 \, dx
$$

$$
= \left[ \frac{x^3}{3} \right]_{1}^{3}
$$

$$
= \frac{3^3}{3} - \frac{1^3}{3} = \frac{26}{3}
$$

2. Evaluate $\int_{1}^{4} x^{-2} \, dx$
$$
\int_{1}^{4} x^{-2} \, dx
$$

$$
= -\left[ \frac{1}{x} \right]_{1}^{4}
$$

$$
= -\frac{1}{4} - \frac{1}{1} = \frac{3}{4}
$$

##### Example 2: Integral of 1
1. Calculate the integral of:
$$
\int_{0}^{5} 1 \, dx
$$

$$
= [{x}]_{0}^{5}
$$

$$
= 5 - 0 = 5
$$

2. Calculate the integral of:
$$
\int_{2}^{7} 1 \, dt
$$

$$
= [{t}]_{2}^{7}
$$

$$
= 7 - 2 = 5
$$

##### Example 3: Integral of $e^x$
1. Calculate the integral of:
$$
\int_{0}^{1} e^x \, dx
$$

$$
= [{e^x}]_{0}^{1}
$$

$$
= e^1 - e^0 = e - 1
$$


##### Example 4: Integral of $a^x$
1. Calculate the integral of:
$$
\int_{0}^{1} 2^x \, dx
$$

$$
= \left[\frac{2^x}{ln \,2} \right]_{0}^{1}
$$

$$
= \frac{2^1}{ln \,2} - \frac{2^0}{ln \,2}
$$

$$
= \frac{2-1}{ln \,2}
$$

$$
= \frac{1}{ln\,2}
$$

2. Calculate integral of:
$$
\int_{1}^{2} 3^x \, dx
$$

$$
= \left[\frac{3^x}{ln \,3} \right]_{1}^{2}
$$

$$
= \frac{3^2}{ln \,3} - \frac{3^1}{ln \,3}
$$

$$
= \frac{9-3}{ln \,3} = \frac{6}{ln \,3}
