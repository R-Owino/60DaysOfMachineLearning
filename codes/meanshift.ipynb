# Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import MeanShift
from sklearn.preprocessing import StandardScaler
```

# Generate random data

```python
np.random.seed(42)
cluster_1 = np.random.randn(100,2) + [17,5]
cluster_2 = np.random.randn(100,2) + [10,17]
cluster_3 = np.random.randn(100,2) + [20,17]
```

```python
cluster_1
```

# Combine the clusters into a dataset by using a v stack

```python
X = np.vstack((cluster_1, cluster_2, cluster_3))
```

# Scaler

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
```

# Apply the mean shift

```python
ms = MeanShift()
ms.fit(x_scaled)
```

# Extraction of the cluster labels

```python
labels = ms.labels_
```

# Visualize the labels

```python
plt.figure(figsize=(10,8))
plt.scatter(x_scaled[:,0], x_scaled[:,1], c=labels, cmap='viridis',s=50)
plt.title("Mean Shift")
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar()
plt.show()
```
